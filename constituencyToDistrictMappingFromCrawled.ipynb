```python
import pandas as pd
```

```python
csr_data_path = '/home/manikaran/M.tech/Minor project/data/CSR_NEW_FILES/2014-15/{}/{}'
```

```python
districts_path = '/home/manikaran/M.tech/Minor project/data/CSR_NEW_FILES/districts_updated.csv'
```

```python
district_to_cons_path = './constituencyCrawl/{}.csv'
```

```python
districtsDf = pd.read_csv(districts_path)
```

```python
i = 0
states_not_done = []
while i < districtsDf.shape[0]:
    state = districtsDf.iloc[i][0]
    try:
        tempDf = pd.read_csv(district_to_cons_path.format(state.replace(' ', '_')))
        temp2Df = pd.read_csv(csr_data_path.format(state, 'Z_total.csv'))
        temp2Df.columns=['CSR_District', 'Amount']
        df = tempDf.merge(temp2Df, left_on='District', right_on='CSR_District', how='left')
        df.to_csv('./constituencyCrawl/WithAmount/2014-15/{}.csv'.format(state))
    except Exception as e:
        print(state, e)
        states_not_done.append(state)
    i += 1
```
