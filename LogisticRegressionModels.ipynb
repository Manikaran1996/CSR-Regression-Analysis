```python
from statsmodels.formula.api import Logit, logit
```

```python
dataDf = pd.read_csv('{}/{}/{}'.format(folder, '2014-15', 'dataForAnalysis.csv'))
```

```python
dataDf[(dataDf['ministry_comp_intersection'] == 1) & (dataDf['invested'] == 1)]
```

```python
dataDf['rs_mplad_dist'] = dataDf['rs_mplad_dist'].fillna(0)
```

```python
reqdCols = ['Type', 'Margin %', 'isMPMinister', 'centerAlignment', 'stateAlignment', 'not_aligned', 'rs_mplad_dist', 'is_backward', 'non_backward', 'ministry_comp_intersection', 'invested']
```

```python
for c in reqdCols:
    print(c, pd.isna(dataDf[c]).sum())
```

```python
df = dataDf[reqdCols].dropna().reset_index(drop=True)
```

```python
df.shape
```

```python
positiveDf = df[df['invested'] == 1]
```

```python
positiveDf.shape
```

```python
negativeDf = df[df['invested'] == 0]
```

```python
sampledEg = negativeDf.sample(n=10000)
```

```python
finalDf = pd.concat((positiveDf, sampledEg), axis=0)
```

```python
finalDf.rename({'Margin %':'Margin'}, axis=1, inplace=True)
```

```python
finalDf.shape
```

```python
finalDf.head()
```

```python
finalDf.reset_index(drop=True, inplace=True)
```

```python
finalDf = pd.concat((pd.get_dummies(finalDf['Type']), finalDf), axis=1).drop('Type', axis=1)
```

```python
model1 = logit("invested ~ GEN:non_backward  + SC:non_backward + ST:non_backward + GEN:is_backward  + SC:is_backward + ST:is_backward", finalDf).fit(maxiter=1000)
model1.summary()
```

```python
model4 = logit("invested ~  GEN:non_backward  + SC:non_backward + ST:non_backward + GEN:is_backward  + SC:is_backward + ST:is_backward + Margin", finalDf).fit(method='cg', maxiter=1000)
model4.summary()
```
