```python
districtsToConstDf = pd.read_csv('{}/districtsWithConstDetails.csv'.format(folder))
districtsToConstDf['State'] = districtsToConstDf['State'].map(str.upper)
districtsToConstDf['District'] = districtsToConstDf['District'].map(str.upper)

#adding ministers info in district to loksabha constituency dataframe
tempDf = pd.DataFrame(ministersDf)
tempDf['LokSabhaConstituency'] = tempDf['LokSabhaConstituency'].map(lambda x: x.upper().strip())
districtsToConstDf = districtsToConstDf.merge(tempDf[['isMPMinister', 'ministry', 'minCode', 'LokSabhaConstituency']], how='left', on='LokSabhaConstituency')
districtsToConstDf['isMPMinister'] = districtsToConstDf['isMPMinister'].fillna(0)
districtsToConstDf.rename(mapper={'ministry':'LokSabha_Ministry', 'minCode':'LokSabha_MinCode'}, axis=1, inplace=True)
districtsToConstDf['LokSabha_MinCode'] = districtsToConstDf['LokSabha_MinCode'].fillna('[]')
districtsToConstDf['LokSabha_Ministry'] = districtsToConstDf['LokSabha_Ministry'].fillna('[]')
districtsToConstDf['Margin %'] = districtsToConstDf['Margin %'].map(lambda x: x.rstrip(' %') if not pd.isna(x) else x)
districtsToConstDf.head()
```

```python
stateDf = pd.read_csv('{}/stateGovernment2014New.csv'.format(folder))
stateDf['State'] = stateDf['State'].apply(lambda x: x.upper().strip())
stateDf.rename(mapper={'PartyShort':'PartyInState'}, axis=1, inplace=True)
stateDf.index = stateDf['State']
stateDf['PartyInState'] = stateDf['PartyInState'].map(eval)
partyInState = stateDf[['PartyInState', 'Party']].to_dict(orient='index')
```

```python
# MPs Alignment with center
centerAlignment = []
stateAlignment = []
for i in range(districtsToConstDf.shape[0]):
    party = districtsToConstDf.iloc[i]['Party']
    partyShort = districtsToConstDf.iloc[i]['PartyShort']
    state = districtsToConstDf.iloc[i]['State']
    if party in parties_in_nda:
        centerAlignment.append(1)
    else:
        centerAlignment.append(0)
    if state not in partyInState:
        print(state)
        stateAlignment.append(0)
        continue
    if partyShort in partyInState[state]['PartyInState']:
        stateAlignment.append(1)
    else:
        stateAlignment.append(0)
districtsToConstDf['centerAlignment'] = centerAlignment
districtsToConstDf['stateAlignment'] = stateAlignment
```

```python
import numpy as np
def getMinistries(sector):
    if pd.isna(sector):
        return []
    tempDf = matrixDf[matrixDf['SectorCode'] == sector]
    #getting column number where values in the row == 1, 
    #since the ministry code starts from 1 and the first column header is SectorCode so
    #the indices return by where function is actually the ministry code
    return list(np.where(np.array(tempDf[:]=='1'))[1])
```

```python
#adding CIN
for fy in ['2014-15']: #, '2015-16', '2016-17']:
    csrDf = pd.read_csv('{}/{}/{}.csv'.format(folder, fy, 'totalCSRAllConstAnalysis1'))
    compWithCINDf = pd.read_csv('{}/all_companies_2014_with_sector.csv'.format(folder))
    compWithCINDf = compWithCINDf[['company', 'CIN', 'buss_sector', 'sectorCode']]
    nic = []
    ministries = []
    year = []
    for i in range(compWithCINDf.shape[0]):
        ministries.append(getMinistries(compWithCINDf.iloc[i]['sectorCode']))
    compWithCINDf['ministriesInfluenced'] = ministries
    #compWithCINDf.to_csv('{}/{}/{}.csv'.format(folder, fy, 'companies'))
    csrDf['company'] = csrDf['company'].map(str.upper, na_action='ignore')
    tempDf = csrDf.merge(compWithCINDf, how='left', on='company')
    print(tempDf['CIN'].isna().sum(), 'None values', tempDf['company'].isna().sum())
    #tempDf.to_csv('{}/{}/{}.csv'.format(folder, fy, 'totalCSRWithComp{}'.format(fy)))
    ministryInfluenceComp = []
    for i in range(tempDf.shape[0]):
        intersection = 0
        minInfluenced = tempDf.iloc[i]['ministriesInfluenced']
        if pd.isna(tempDf.iloc[i]['CIN']):
            ministryInfluenceComp.append(0)
            continue
        if tempDf.iloc[i]['isMPMinister'] == 1:
            mins = eval(tempDf.iloc[i]['LokSabha_MinCode'])
            for ministry in minInfluenced:
                if ministry in mins:
                    intersection = 1
        if tempDf.iloc[i]['rs_min_mplad_district'] == 1:
            mins = eval(tempDf.iloc[i]['rs_min_ministry'])
            for ministry in minInfluenced:
                if ministry in mins:
                    intersection = 1
        ministryInfluenceComp.append(intersection)
    tempDf['ministry_comp_intersection'] = ministryInfluenceComp
    tempDf.to_csv('{}/{}/{}.csv'.format(folder, fy, 'totalCSRAllConstLinearReg'))
```

```python
def Or(series):
    ans = 0
    for s in series:
        ans = ans or s
    return ans
```

```python
def Sum(series):
    total = 0
    for x in series:
        x = int(x.replace(',','').replace('.',''))
        total += x
    return total
```

```python
fy = '2014-15'
totalCSRDf = pd.read_csv('{}/{}/totalCSRAllConstLinearReg.csv'.format(folder, fy))
reqdCols = ['Constituency', 'state', 'district', 'Type', 'Margin %', 'isMPMinister', 'centerAlignment', 'stateAlignment', 'rs_min_mplad_district', 'is_backward', 'ministry_comp_intersection', 'amount']
dataDf = totalCSRDf[reqdCols]
groupedDf = dataDf.groupby(['Constituency', 'state', 'Type', 'Margin %', 'isMPMinister', 'centerAlignment', 'stateAlignment'])
```

```python
#added ministry codes to the linearRegression analysis final data
def concat(series):
    res = []
    for x in series:
        res += eval(x)
    return list(set(res))
fy='2014-15'
totalCSRDf = pd.read_csv('{}/{}/totalCSRAllConstLinearReg.csv'.format(folder, fy))
reqdCols2 = ['Constituency', 'state', 'district', 'Type', 'Margin %', 'isMPMinister', 'centerAlignment', 'stateAlignment', 'rs_min_mplad_district', 'is_backward', 'ministry_comp_intersection', 'LokSabha_MinCode', 'rs_min_ministry', 'amount']
dataDf2 = totalCSRDf[reqdCols2]
groupedDf2 = dataDf2.groupby(['Constituency', 'state', 'Type', 'Margin %', 'isMPMinister', 'centerAlignment', 'stateAlignment'])
tempDf2 = groupedDf2.agg({'rs_min_mplad_district':Or, 'is_backward':Or, 'ministry_comp_intersection':Or, 'amount':Sum, 'LokSabha_MinCode':concat, 'rs_min_ministry':concat}).reset_index()
tempDf2["non_backward"] = tempDf2["is_backward"].apply(lambda x: 0 if x == 1 else 1)
not_aligned = []
for i in range(tempDf2.shape[0]):
    not_aligned.append(1-(tempDf2.iloc[i]["stateAlignment"] or tempDf2.iloc[i]["centerAlignment"]))
tempDf2["not_aligned"] = not_aligned
tempDf2.to_csv('{}/{}/dataForLinearRegressionAnalysis2.csv'.format(folder, '2014-15'))
```
