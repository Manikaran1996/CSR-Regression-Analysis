```python
CSRDistrictsDf = pd.read_csv('/home/manikaran/M.tech/Minor project/data/CSR_Districts.csv')
```

```python
mergedDf = mergedDf.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis=1)
```

```python
CSR_Districts = []
for i in range(mergedDf.shape[0]):
    if pd.isna(mergedDf.iloc[i]['District']):
        CSR_Districts.append(None)
        continue
    state = mergedDf.iloc[i]['State']
    aliases = mergedDf.iloc[i]['DistrictAliasName']
    tempDf = CSRDistrictsDf[CSRDistrictsDf['State']==state]
    districts = []
    for j in range(len(aliases)):
        alias = aliases[j]
        for k in range(tempDf.shape[0]):
            if tempDf.iloc[k]['District'].strip() in alias:
                districts.append(tempDf.iloc[k]['District'])
                break
        else:
            if 'Bardhaman' in alias:
                districts.extend(['Purba Bardhaman', 'Paschim Bardhaman']) #since bardhaman is bifurcated and Bardhaman also exists in database
            else:    
                districts.append(None)
    CSR_Districts.append(districts)
```

```python
mergedDf['CSR_District'] = CSR_Districts
```

```python
CSR_file = '/home/manikaran/M.tech/Minor project/data/CSR_NEW_FILES/{}/{}/Z_total.csv'
```

```python
years = ['2014-15', '2015-16', '2016-17']
for fy in years:    
    amount = []
    for i in range(mergedDf.shape[0]):
        state = mergedDf.iloc[i]['State']
        districts = mergedDf.iloc[i]['CSR_District']
        if districts == None:
            amount.append(None)
            continue
        s = 0
        for d in districts:
            try:
                tempDf = pd.read_csv(CSR_file.format(fy,state))
                tempDf.columns = ['district', 'amount']
                #print(state,districts,tempDf[tempDf['district'] == d]['amount'])
                if pd.isna((tempDf[tempDf['district'] == d]['amount']).iloc[0]):
                    continue
                s += (tempDf[tempDf['district'] == d]['amount'].iloc[0])
            except Exception as e:
                print(e)
                print(d, tempDf)
        amount.append(s)
    mergedDf['CSR_Amount_' + fy] = amount
```
